#### Import necessary libraries


```python
from imageai.Detection import ObjectDetection
import cv2 as cv
import matplotlib.pyplot as plt
```

#### Initialize the ObjectDetection instance


```python
detector = ObjectDetection()
```

#### Set the model type to YOLOv3


```python
detector.setModelTypeAsYOLOv3()
```

#### Specify the path to the YOLOv3 model file


```python
detector.setModelPath('data/yolov3.pt')
```

#### Load the YOLOv3 model


```python
detector.loadModel()
```

#### Perform object detection on the input image, setting a minimum detection probability of 70%


```python
detection = detector.detectObjectsFromImage(input_image='data/tehran.png', 
                                           output_image_path='tehran-detected.jpg', 
                                           minimum_percentage_probability=70)
```

#### Display the detected objects in the image


```python
image_detected = cv.imread('tehran-detected.jpg')
plt.imshow(cv.cvtColor(image_detected, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
